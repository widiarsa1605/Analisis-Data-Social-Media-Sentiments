```python
import pandas as pd # Untuk visualisasi data
from sklearn.feature_extraction.text import CountVectorizer # Untuk Bag of Words
import spacy # Untuk Part Of Speech Taging

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, chi2
```

## Baca Data

```python
data = pd.read_csv('data/sentimentdataset.csv')
data_sentimen = data[['Text', 'Sentiment']]

def remove_spaces(data):
    return data.strip()
data_sentimen['Text'] = data_sentimen['Text'].apply(remove_spaces)
data_sentimen['Sentiment'] = data_sentimen['Sentiment'].apply(remove_spaces)


data_sentimen.head()
```

## Bag of Words

```python
vectorizer = CountVectorizer()

# data ke Bag of Words
bow_data = vectorizer.fit_transform(data_sentimen['Text'])

# Konversi ke DataFrame

bow_df = pd.DataFrame(bow_data.toarray(), columns=vectorizer.get_feature_names_out())

bow_df.head()
```

## Part Of Speech Taging

```python
nlp = spacy.load('en_core_web_sm') # Model spacy

def part_of_speech(text):
    doc = nlp(text)
    doc_pos =  [(token.text, token.pos_) for token in doc]
    return ' '.join([tag for _, tag in doc_pos])


data_sentimen['POS_Tagging'] = data_sentimen['Text'].apply(part_of_speech)

data_sentimen.head()
# data_sentimen.shape
```

## Seleksi Fitur

```python
data_sentimen['Sentiment'] = pd.factorize(data_sentimen['Sentiment'])[0] #Ekstraksi Fitur

# Fungsi untuk transformasi POS tagging
def pos_transformer(X):
    # Menghitung frekuensi POS Tag
    pos_counts = X.apply(lambda x: pd.Series(x.split()).value_counts()).fillna(0)
    return pos_counts

# Column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('text', TfidfVectorizer(), 'Text'),
        ('pos', FunctionTransformer(func=pos_transformer, validate=False), 'POS_Tagging')
    ],
    remainder='passthrough'
)

# Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectKBest(chi2, k=10))
])

# Data
X = data_sentimen[['Text', 'POS_Tagging']]
y = data_sentimen['Sentiment']

# Transformasi fitur
X_transformed = pipeline.fit_transform(X, y)

# Menampilkan fitur yang dipilih
# Mendapatkan nama fitur dari masing-masing transformer
text_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['text'].get_feature_names_out()
pos_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['pos'].transform(X['POS_Tagging']).columns.tolist()

# Menggabungkan nama fitur
feature_names = text_feature_names.tolist() + pos_feature_names

# Mengambil indeks fitur yang dipilih
selected_feature_indices = pipeline.named_steps['feature_selection'].get_support(indices=True)
selected_features = [feature_names[i] for i in selected_feature_indices]

print("Fitur yang dipilih:")
print(selected_features)
```

## Reduksi Dimensi

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('text', TfidfVectorizer(), 'Text'),
        ('pos', OneHotEncoder(), 'POS_Tagging')
    ],
    remainder='passthrough'
)

# Langkah 2: Pipeline dengan penggabungan fitur dan PCA
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectKBest(chi2, k=10)),  # Pilih 10 fitur terbaik
    ('scaler', StandardScaler(with_mean=False)),  # Standardisasi data
    ('pca', PCA(n_components=5))   # Reduksi dimensi menjadi 5 komponen utama
])
```
